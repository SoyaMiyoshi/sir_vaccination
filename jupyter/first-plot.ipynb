```python
from matplotlib.pyplot import subplots
import numpy as np

def my_plot(df, my_label, my_color):
    axarr[0].plot(df['time'], df['avg outbreak size'], c=my_color, label=my_label)
    axarr[0].set_xlabel("SEASON")
    axarr[0].set_ylabel("AVG OUTBREAK SIZE")

    axarr[1].plot(df['time'], df['avg time to extinction'], c=my_color)
    axarr[1].set_xlabel("SEASON")
    axarr[1].set_ylabel("AVG TIME TO EXTINCTION")

    axarr[2].plot(df['time'], df['coverage'], c=my_color)
    axarr[2].set_xlabel("SEASON")
    axarr[2].set_ylabel("COVERAGE")
    
```

```python
path = '../output'
import os
import re
import pandas as pd

def generate_regex(query):
    return(query + '=(.*)[0-9]')

def find_contents(query, search_in):
    pattern = re.compile(r"%s"%generate_regex(query))
    match = pattern.search(search_in)
    if match:
        return(match.group(0))
    
directory = os.fsencode(path)

f, axarr = subplots(1, 3, figsize=(25,5)) 
count = 0
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    name = find_contents("z", filename)
    if(name):
        p = path + "/" + os.fsdecode(filename)
        df = pd.read_csv(p)
        color = "C" + str(count)
        my_plot(df, name, color)
        count += 1

f.legend()

        
```
